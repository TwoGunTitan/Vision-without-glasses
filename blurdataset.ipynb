```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
import os
```

```python
image = cv2.imread("/content/drive/MyDrive/dogs_vs_cats_dataset/test_data/normal_dataset/cat.110.jpg")[...,::-1]
```

```python
def plot_images(img1, img2, title1="", title2=""):
    fig = plt.figure(figsize=[15,15])
    ax1 = fig.add_subplot(121)
    ax1.imshow(img1)
    ax1.set(xticks=[], yticks=[], title=title1)

    ax2 = fig.add_subplot(122)
    ax2.imshow(img2)
    ax2.set(xticks=[], yticks=[], title=title2)
```

```python
sigma = 4
copy = cv2.blur(image, (sigma*4+1,sigma*4+1))
plot_images(image,copy)
```

```python
sigmas = [5,7,8,10]
objects = ['cat.' ,'dog.']
for sigma in sigmas:
  directory = r'/content/drive/MyDrive/dogs_vs_cats_dataset/test_data/gaussian_'+str(sigma)
  os.chdir(directory)
  for i in range(2500):
    for object in objects:
      image = cv2.imread("/content/drive/MyDrive/dogs_vs_cats_dataset/test_data/normal_dataset/"+object+str(i)+".jpg")[...,::-1]
      blurred_image = cv2.GaussianBlur(image, (sigma*6+1,sigma*6+1),sigma)
      filename = object+str(i)+'.jpg'
      blurred_image = blurred_image[...,::-1]
      cv2.imwrite(filename,blurred_image)
```
